```python
# import matplotlib.pyplot as plt
# import numpy as np

# last_week_cups = (20, 35, 30, 35, 27)
# this_week_cups = (25, 32, 34, 20, 25)
# names = ['Mary', 'Paul', 'Billy', 'Franka', 'Stephan']

fig = plt.figure(figsize=(4,6), dpi=200)
left, bottom, width, height = 0.1, 0.3, 0.8, 0.6
ax = fig.add_axes([left, bottom, width, height]) 
 
width = 0.35   
ticks = np.arange(len(df['activity']))    
ax.bar(ticks, df['consumption'], width, label='consumption', color='blue')
ax.bar(ticks + width, df['recommanded_quantity'], width, align="center",
    label='recommanded_quantity',color='orange')

ax.set_ylabel('Quantity (L)')
ax.set_xlabel('Activity')
ax.set_title('Usages')
ax.set_xticks(ticks + width/2)
ax.set_xticklabels(df['activity'])

ax.legend(loc='best')
plt.show()
```

```python
df1 = pd.read_excel("profil.xlsx")
df1
```

```python
df2 = df1.groupby('profil', as_index=False)['consumption'].sum()
df2
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [8, 8]  
# Setting labels for items in Chart
Employee = df2['profil']
  
# Setting size in Chart based on 
# given values
# Salary = [40000, 50000, 70000, 54000, 44000]
  
# colors
colors = ['#FF0000', '#0000FF', '#FFFF00', 
          '#ADFF2F', '#FFA500']
# explosion
explode = (0.05, 0.05, 0.05, 0.05)
  
# Pie Chart
plt.pie(df2['consumption'], colors=colors, labels=Employee,
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)
  
# draw circle
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
  
# Adding Title of chart
plt.title('consumption per profil')
  
# Displaying Chart
plt.show()
```
